```python
import numpy as np
from windfarm import Windfarm
from matrix_operators import crossover, inversion, mutation
```

```python
N = 10
R = 40
alpha = 0.5/np.log(60/0.3)
C_t = 0.88
```

```python
placement_1 = np.full((N, N), False)
for i in range(N):
  for j in range(i, N):
    placement_1[i, j] = True

placement_2 = placement_1.T
```

```python
# Input placements for the examples
Windfarm(placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
Windfarm(placement_2, 10 * R, 10 * R, R, alpha, C_t).show()
```

```python
# Example of crossover with vertical cut
new_placement_1, new_placement_2 = crossover(placement_1, placement_2)
Windfarm(new_placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
Windfarm(new_placement_2, 10 * R, 10 * R, R, alpha, C_t).show()
```

```python
# Example of crossover with horizontal cut
new_placement_1, new_placement_2 = crossover(placement_1, placement_2, vertical = False)
Windfarm(new_placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
Windfarm(new_placement_2, 10 * R, 10 * R, R, alpha, C_t).show()
```

```python
# Example of inversion with vertical cuts
new_placement_1 = inversion(placement_1)
Windfarm(new_placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
```

```python
# Example of inversion with horizontal cuts
new_placement_1 = inversion(placement_1, vertical = False)
Windfarm(new_placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
```

```python
# Example of mutation with probability of 10%
new_placement_1 = mutation(placement_1, 0.1)
Windfarm(new_placement_1, 10 * R, 10 * R, R, alpha, C_t).show()
```
